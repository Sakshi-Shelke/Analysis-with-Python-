## 2. Data Cleaning
- Data cleaning is a crucial step that involves addressing any missing, inconsistent, or erroneous data.
- In this project, several columns contained missing values, especially the 'Warehouse Name'and 'Product Colour' columns.
- These missing values were handled through various techniques, such as filling missing entries with average values (for numerical columns) or with placeholders (for categorical columns).
- Additionally, the 'Warehouse Name' was mapped to the corresponding city based on the 'Country' column to standardize the data.

```python
df['Country'].fillna(df['Country'].mode()[0],inplace = True)
df.head(20)
```

```python
df['Warehouse name'] = None  # This will empty the column and make all values NaN
df.head()
```

```python
# Define the country-to-warehouse mapping
country_warehouse_map = {
    'Japan': 'Tokyo',
    'UK': 'London',
    'France': 'Paris',
    'USA': 'New York',
    'Italy': 'Milan',
    'Germany': 'Berlin'
}
# Function to impute warehouse name based on country
def impute_warehouse_name(row):
    if pd.isna(row['Warehouse name']):  # Check if Warehouse Name is missing
        return country_warehouse_map.get(row['Country'])  # Impute based on country
    return row['Warehouse name']
# Apply the function to impute missing Warehouse Name values
df['Warehouse name'] = df.apply(impute_warehouse_name, axis=1)
# Save the updated DataFrame to CSV file
df.to_csv('Luxury Store Unclean Data.csv', index=False)
df.head(10)
```

```python
# Define a mapping function for imputing missing values in 'Product Category'
def impute_product_category(row):
    if pd.isna(row['Product Category']):
        gender = row['Product Gender Target']
        ptype = row['Product Type']
        if gender == 'Male' and ptype == 'Shoes':
            return 'men shoes'
        elif gender == 'Female' and ptype == 'Shoes':
            return 'women shoes'
        elif gender == 'Male' and ptype == 'Clothing':
            return 'men clothing'
        elif gender == 'Female' and ptype == 'Clothing':
            return 'women clothing'
        elif gender == 'Male' and ptype == 'Accessories':
            return 'men accessories'
        elif gender == 'Female' and ptype == 'Accessories':
            return 'women accessories'
        elif gender == 'Unisex' and ptype == 'Accessories':
            return 'unisex accessories'
        elif gender == 'Unisex' and ptype == 'Clothing':
            return 'unisex clothing'
        elif gender == 'Unisex' and ptype == 'ShMaleoes':
            return 'unisex shoes'
    return row['Product Category']

# Apply the mapping function row-wise
df['Product Category'] = df.apply(impute_product_category, axis=1)

# Save the updated DataFrame back to the same CSV file
df.to_csv('Luxury Store Unclean Data.csv', index=False)
```

## 4. Visualization
- Visualization is a powerful tool for communicating insights derived from data. In this project, we created various visualizations to better understand the data:
    - Histograms for the price distribution of products, highlighting trends and outliers.
    - Count plots to show the distribution of product categories, seasons, conditions, and sales by gender.
    - Boxplots to visualize the price distribution across different product categories and seasons, helping identify patterns and outliers.
    - Correlation heatmaps to understand relationships between numerical variables. These visualizations help in presenting the data in an intuitive way, enabling better decision-making.

```python
# Group data by 'Product Gender Target' and calculate counts
gender_distribution = df['Product Gender Target'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    gender_distribution, 
    labels=gender_distribution.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('coolwarm', n_colors=3),  # Customize colors
    explode=[0.05 if i == gender_distribution.idxmax() else 0 for i in gender_distribution.index]  # Highlight the largest slice
)
plt.title('Gender Distribution of Products')
plt.show()
```

#### Insights
- The Unisex Type of Product Target Gender comprises of the highest number from the total counts of products.
- Although the difference is not very varied from the other Product Target Gender type, thus distribution is almost even.

```python
# Calculate revenue contribution by brand
revenue_by_brand = df.groupby('Brand Name')['Price Usd'].sum().sort_values(ascending=False).head(5)

# Plot
plt.figure(figsize=(8, 8))
plt.pie(
    revenue_by_brand, 
    labels=revenue_by_brand.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('coolwarm', n_colors=5),
    explode=[0.05 if i == revenue_by_brand.idxmax() else 0 for i in revenue_by_brand.index]  # Highlight the largest slice
)
plt.title('Top 5 Brands by Revenue Contribution')
plt.show()
```

```python
# Group data for average prices
category_price = df.groupby('Product Category')['Price Usd'].mean()

# Cumulative trend
cumulative_price = category_price.cumsum()

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar graph for average price
category_price.plot(kind='bar', color='brown', alpha=0.6, ax=ax1, label='Average Price')
ax1.set_ylabel('Average Price (USD)')
ax1.set_title('Average Price by Product Category with Cumulative Trend')

# Line graph for cumulative price
ax2 = ax1.twinx()
ax2.plot(cumulative_price, color='red', marker='o', linestyle='-', label='Cumulative Price Trend')
ax2.set_ylabel('Cumulative Price (USD)')
ax2.legend(loc='upper left')

plt.show()
```

#### Insights
- Average Prices:
    - "Unisex clothing" and "women shoes" have the highest average prices.
    - "Men accessories" and "men clothing" have relatively lower average prices.
- Cumulative Trend:
    - The cumulative price trend (red line) steadily increases as categories are added, indicating their cumulative contribution to overall pricing.
    - "Women shoes" contribute significantly to the final cumulative total.
- Pricing Focus:
    - High-average categories like "women shoes" and "unisex clothing" may require targeted pricing strategies or promotions to maximize profitability.

```python
# Group data by season for sold products
season_sales = df[df['Sold'] == True].groupby('Product Season').size()

# Cumulative trend
cumulative_sales = season_sales.cumsum()

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar graph for sales by season
season_sales.plot(kind='bar', color='brown', alpha=0.8, ax=ax1, label='Products Sold')
ax1.set_ylabel('Number of Products Sold')
ax1.set_title('Products Sold by Season with Cumulative Trend')

# Line graph for cumulative sales
ax2 = ax1.twinx()
ax2.plot(cumulative_sales, color='blue', marker='o', linestyle='-', label='Cumulative Sales Trend')
ax2.set_ylabel('Cumulative Sales')
ax2.legend(loc='upper left')

plt.show()
```

#### Insights
- Sales by Season:
    - Sales are highest in Winter, followed by Summer and Spring.
    - Autumn has the lowest number of products sold.
- Cumulative Trend:
    - The cumulative sales trend (blue line) rises steadily, with a sharp increase by Winter, highlighting its significant contribution to total sales.
- Seasonal Focus:
    - Winter is a critical sales period, suggesting a focus on marketing and inventory during this season.
    - Autumn may require strategies to boost sales, such as discounts or targeted promotions.

```python
# Group data for bar heights
data = df.groupby(['In Stock', 'Sold']).size().unstack()

# Bar graph with line overlay
fig, ax1 = plt.subplots(figsize=(8, 5))

# Bar graph for counts
data.plot(kind='bar', ax=ax1, alpha=0.7, color=['brown', 'lightblue'])
ax1.set_ylabel('Count')
ax1.set_title('In Stock vs Sold: Count and Trend')

# Line graph for cumulative trend (example: total count)
cumulative = data.sum(axis=1).cumsum()
ax2 = ax1.twinx()
ax2.plot(cumulative, color='red', marker='o', label='Cumulative Trend')
ax2.set_ylabel('Cumulative Count')
ax2.legend(loc='upper left')

plt.show()
```

#### Insights
- This graph shows the count of items that are "In Stock" (True) versus those that are "Sold" (False). It also includes a line graph showing the cumulative trend of the total number of items over time.
- More items are currently "In Stock" than have been "Sold."
- The cumulative trend indicates a steady increase in the total number of items.

```python
# Calculate proportions
sold_in_stock = df.groupby(['In Stock', 'Sold']).size().unstack()
print(sold_in_stock)

# Heatmap for Proportions
sns.heatmap(sold_in_stock, annot=True, fmt='d', cmap='coolwarm')
plt.title('Proportion of Products: In Stock vs Sold')
plt.show()
```

#### Insights
- The heatmap shows the proportion of products that are "In Stock" (True) versus those that are "Sold" (False).
- A significantly higher proportion of products are "In Stock" compared to those that are "Sold."
